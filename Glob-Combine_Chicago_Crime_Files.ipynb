# Using Glob to find all files to then combine them 

```python
import pandas as pd
import glob
# path for files
q = "Data/Chicago/Chicago-Crime_*.csv"
# chunk files as sorted
chunked_files = sorted(glob.glob(q))
# view files found
chunked_files
```

```python
## Loading and Concatenating the list of dfs with 1 line
df_combined = pd.concat([pd.read_csv(file, index_col=0) for file in chunked_files])
df_combined
```

```python
## Saving the final combined dataframe
final_fname ='Data/Chicago/crime_data_combined.csv.gz'
df_combined.to_csv(final_fname, compression='gzip', index=False)
```

```python
# Load in combined CSV to confirm
df = pd.read_csv('Data/Chicago/crime_data_combined.csv.gz',lineterminator='\n')
df.head()
```
